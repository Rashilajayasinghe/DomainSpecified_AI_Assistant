```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
data = pd.read_csv('/content/sample_data/fin_synthetic_samples_cleaned (4).csv')
```

```python
data
```

```python
grade_to_points = {
    'A+': 4.0,
    'A' : 4.0,
    'A-': 3.7,
    'B+': 3.3,
    'B' : 3.0,
    'B-': 2.7,
    'C+': 2.3,
    'C' : 2.0,
    'C-': 1.7,
    'D+': 1.3,
    'D' : 1.0,
    'E' : 0.0
}
```

# Task
Identify all columns in the `data` DataFrame that start with 'Grade_'. Then, iterate through these columns and map the letter grades to numerical point values using the `grade_to_points` dictionary, assigning `NaN` for 'NOT_TAKEN' grades. Finally, display a sample of the updated `data` DataFrame to show the new grade point columns.

## Identify Grade Columns

### Subtask:
Identify all columns in the `data` DataFrame that start with 'Grade_'. These are the columns for which we need to convert grades to points.


**Reasoning**:
To identify all columns that start with 'Grade_', I will iterate through the column names of the `data` DataFrame and filter them using a string method. These columns will be stored in a list.



```python
grade_columns = [col for col in data.columns if col.startswith('Grade_')]
print(f"Identified grade columns: {grade_columns}")
```

## Convert Grades to Numerical Points

### Subtask:
Iterate through the identified 'Grade_' columns and convert the letter grades to numerical point values using the `grade_to_points` dictionary. Assign `NaN` for 'NOT_TAKEN' grades.

#### Instructions
1. For each column in `grade_columns`, apply a mapping function.
2. The mapping function should use the `grade_to_points` dictionary to convert letter grades to numerical values.
3. If a grade is 'NOT_TAKEN', convert it to `np.nan`.

**Reasoning**:
To convert the letter grades in the identified columns to numerical point values, I will iterate through each grade column and apply a mapping function that uses the `grade_to_points` dictionary. Grades marked 'NOT_TAKEN' will be converted to `np.nan`.



```python
for col in grade_columns:
    data[col] = data[col].replace('NOT_TAKEN', np.nan).map(grade_to_points)
print("Grades successfully converted to numerical points.")
```

```python
print("Sample of the updated DataFrame with converted grade points:")
data[grade_columns].head()
```
